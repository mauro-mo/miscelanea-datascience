```python
#creamos una serie de caracteres
#el tipo es objeto

cad = pd.Series( ['h', 'o', 'l', 'o', 'g', 'r', 'a', 'm'] )
cad
```

```python
#creamos un DataFrame que es una estructura que se aplica mucho en ML y EDA.

lst = ['Hola', 'mundo', 'robótico']
df = pd.DataFrame(lst)
print(df)
```

```python
#creamos una estructura con una cabecera

data = { 'Nombre': ['Juan', 'Ana', 'Jose', 'Arturo'], 
        'Edad': [25, 18, 23, 27],
        'Pais': ['MX', 'CO', 'BR', 'ARG']}

df = pd.DataFrame(data)
print(df)
```

```python
type(df)
```

```python
#imprimimos dos columnas de esta estructura

print(df[['Nombre', 'Pais']])
```
